<h1 style="font-size:1rem;color:blue;">Visualizion

```python
import matplotlib.pyplot as plt
%matplotlib notebook
data = pd.read_csv('cleaned_df.csv')
```

```python
### Line Charts
```

```python

plt.figure(figsize=(10,5))
plt.plot(data.Date, data.ConfirmedCovidCases)
plt.title('Daily Confirmed Cases Covid-19 Ireland 2020-2022')
plt.ylabel('Cases Per 1000x')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.plot(data.Date, data.TotalConfirmedCovidCases)
plt.title('Total Confirmed Cases Covid-19 Ireland 2020-2022')
plt.ylabel('Cases Per Mil')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.Date, data.Dose1Cum, label='1/2 Vaccines')
plt.plot(data.Date, data.Dose2Cum, label='2/2 Vaccines')
plt.plot(data.Date, data.SingleDoseCum, label='Single Dose Vaccines')
plt.plot(data.Date, data.AdditionalDoseCum, label='Booster')
plt.title('Cumulative figures on the roll out of vaccines in Ireland 2020-2022')
plt.ylabel('Cases Per Mil')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.Date, data.HospitalisedCovidCases, label='Hospitalised Covid Cases')
plt.title('Cumulative figures on the Covid Hospitizations in Ireland 2020-2022')
plt.ylabel('Cases Per 1000x')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.Date, data.RequiringICUCovidCases, label='Hospitalised Covid Cases Requiring ICU')
plt.title('Cumulative figures on the Covid Hospitizations requiring ICU in Ireland 2020-2022')
plt.ylabel('Cases Per 100x')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.legend()
plt.show()
```

```python
### Bar Chart
```

```python
plt.figure(figsize=(10,5))
plt.bar(data.Date, data.ConfirmedCovidDeaths)
plt.title('Daily COVID-19 Deaths in Ireland 2020-2022')
plt.ylabel('Deaths')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data.Date, data.TotalCovidDeaths)
plt.title('Total COVID-19 Deaths in Ireland 2020-2022')
plt.ylabel('Cases Per 1000x')
plt.xlabel('Date')
plt.xticks(data.Date[::100])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data.Date, data.Dose1, label='1/2 Vaccines')
plt.bar(data.Date, data.Dose2, label='2/2 Vaccines')
plt.bar(data.Date, data.SingleDose, label='Single dose vaccines')
plt.bar(data.Date, data.AdditionalDose, label='Booster')
plt.title('Daily Vaccine Rollout in Ireland 2020-2022')
plt.ylabel('Cases Per 1000x')
plt.xlabel('Date')
plt.legend()
plt.xticks(data.Date[::100])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
target_row_index = 751
plt.figure(figsize=(10,10))
values = df2.iloc[target_row_index, 1:]
labels = df2.columns[1:]
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
plt.title('Covid Infections in relation to Gender')

plt.pie(values, labels=labels, autopct=make_autopct(values))
plt.show()
```
